# Lets Clean

```python
def preprocess(df):
    for label,content in df.items():
        if pd.api.types.is_numeric_dtype(content) or pd.api.types.is_float_dtype(content):
            if pd.isnull(content).sum():
                df[label]=content.fillna(content.mean())
        if not pd.api.types.is_numeric_dtype(content):
            df[label]=content.astype('category').cat.as_ordered()
            df[label]=pd.Categorical(content).codes+1
    return df
```

# Modelling
We will Try to work with Two Regression Models
1. GradientBoostingRegression
2. RandomForestRegressor

```python
from sklearn.model_selection import train_test_split
np.random.seed(42)
x=df.drop('SalePrice',axis=1)
y=df['SalePrice']
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2)
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor
np.random.seed(42)
g_model_base=GradientBoostingRegressor(n_estimators=155)
g_model_base.fit(x_train,y_train)
g_model_base.score(x_val,y_val)
```

```python
r_model_base=RandomForestRegressor()
r_model_base.fit(x_train,y_train)
r_model_base.score(x_val,y_val)
```

```python
#For GradientBoostingRegressor
np.random.seed(42)
for i in range(100,200,5):
    clf=GradientBoostingRegressor(n_estimators=i)
    clf.fit(x_train,y_train)
    print(f"for n_estimator {i} R2 = {clf.score(x_val,y_val)*100:0.2f}")
```

```python
#For RandomForestRegressor
np.random.seed(42)
for i in range(100,200,5):
    clf=RandomForestRegressor(n_estimators=i)
    clf.fit(x_train,y_train)
    print(f"for n_estimator {i} R2 = {clf.score(x_val,y_val)*100:0.2f}")
```

```python
#Linear
from sklearn.linear_model import LinearRegression
l_model_base=RandomForestRegressor()
l_model_base.fit(x_train,y_train)
l_model_base.score(x_val,y_val)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
def evaluate(y_test,y_pred):
    print('R2 Score = ', r2_score(y_test,y_pred))
    print('Mean absolute Error = ',mean_absolute_error(y_test,y_pred))
    print('Mean squared Error = ',mean_squared_error(y_test,y_pred))
    print('RMSE = ',np.sqrt(mean_squared_error(y_test,y_pred)))
y_pred=g_model_base.predict(x_val)
v=pd.DataFrame({'Actual':y_val,
                'Predict':y_pred
               })
v.head(20)
```

```python
grid={'loss':['squared_error', 'absolute_error', 'huber', 'quantile'],
     'criterion':['friedman_mse', 'squared_error'],
      'max_features':[1, 'sqrt', 'log2'],
      'n_estimators':[180]
     }
from sklearn.model_selection import GridSearchCV
g_modified=GridSearchCV(estimator=GradientBoostingRegressor(),
                       param_grid=grid,
                        verbose=True,
                       cv=5)
g_modified.fit(x_train,y_train)
```
